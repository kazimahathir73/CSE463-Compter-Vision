```python
def transform_5(img_link):
    img = cv2.imread(img_link)
    r, c = img.shape[0], img.shape[1]
    img_crop = img[:r//2, :c//2]                                     # croping th image with half from row and half from column
    img_flip = cv2.flip(img, 1)                                           # horizontal flip
    img_resize = cv2.resize(img, (int(c*0.4), int(r*0.3)))                         # making the image 70% smaller using resize
    trans_arr = np.float32([[1, 0, 100], [0, 1, 100]])
    img_translated = cv2.warpAffine(img, trans_arr, (c, r))                           # translating the image with 100 pixel from x and y axis both
    shear_arr = np.float32([[1, 0.3, 0],[0.3, 1, 0]])
    img_sheared = cv2.warpAffine(img, shear_arr, (c + int(0.3 * r), r + int(0.3 * c)))           # shearing the image with 30% from x and y axis both

    _, (ax1,ax2,ax3,ax4,ax5,ax6) = plt.subplots(1,6,figsize=(20,8))
    ax1.set_title('BGR')
    img = img.astype('uint8')
    ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    ax2.set_title('Crop')
    img_crop = img_crop.astype('uint8')
    ax2.imshow(cv2.cvtColor(img_crop, cv2.COLOR_BGR2RGB))
    ax3.set_title('Flip (Horizontal)')
    img_flip = img_flip.astype('uint8')
    ax3.imshow(cv2.cvtColor(img_flip, cv2.COLOR_BGR2RGB))
    ax4.set_title('Resize')
    img_resize = img_resize.astype('uint8')
    ax4.imshow(cv2.cvtColor(img_resize, cv2.COLOR_BGR2RGB))
    ax5.set_title('Translation')
    img_translated = img_translated.astype('uint8')
    ax5.imshow(cv2.cvtColor(img_translated, cv2.COLOR_BGR2RGB))
    ax6.set_title('Shear')
    img_sheared = img_sheared.astype('uint8')
    ax6.imshow(cv2.cvtColor(img_sheared, cv2.COLOR_BGR2RGB))
```

```python
def random_noise(img_link):
    img = cv2.imread(img_link)
    noisy_img = np.copy(img)

    size = img.size

    coordinates = []

    n_salt = np.ceil(0.5 * size * 0.5).astype(int)                     # Salt noise (white pixels)
    for i in img.shape:
        coordinate = np.random.randint(0, i - 1, n_salt)
        coordinates.append(coordinate)

    noisy_img[coordinates[0], coordinates[1]] = 255

    n_pepper = np.ceil(0.5 * size * 0.5).astype(int)                     # Pepper noise (black pixels)
    for i in img.shape:
        coord = np.random.randint(0, i - 1, n_pepper)
        coordinates.append(coordinate)

    noisy_img[coordinates[0], coordinates[1]] = 0

    noise = np.concatenate(coordinates).flatten()               # flattern the x, y coordinates or 2d matrix to 1d

    _, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(30,8))
    ax1.set_title('BGR')
    img = img.astype('uint8')
    ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

    ax2.set_title('Noisy Image')
    noisy_img = noisy_img.astype('uint8')
    ax2.imshow(cv2.cvtColor(noisy_img, cv2.COLOR_BGR2RGB))

    ax3.hist(noise, bins=50, density=True, alpha=0.6, color='g')
    ax3.set_title('Generated Random Noise')
    ax3.set_xlabel('Value')
    ax3.set_ylabel('Frequency')

    plt.show()
```

```python
transform_5('/content/img1.jpg')
random_noise('/content/img1.jpg')
```
