```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread('/content/img11.jpg')
img2 = cv2.imread('/content/img12.jpg')
img3 = cv2.imread('/content/img13.jpg')
img4 = cv2.imread('/content/img14.jpg')
img5 = cv2.imread('/content/img15.jpg')
```

```python
print(f'Image1 shape - {img1.shape}')
print(f'Image2 shape - {img2.shape}')
print(f'Image3 shape - {img3.shape}')
print(f'Image4 shape - {img4.shape}')
print(f'Image5 shape - {img5.shape}')
```

```python
img2 = cv2.resize(img2, (600, 900))
img4 = cv2.resize(img4, (600, 900))
img5 = cv2.resize(img5, (600, 900))
```

```python
print(f'Image1 shape - {img1.shape}')
print(f'Image2 shape - {img2.shape}')
print(f'Image3 shape - {img3.shape}')
print(f'Image4 shape - {img4.shape}')
print(f'Image5 shape - {img5.shape}')
```

```python
img_p = (0.2*img1)+(0.2*img2)+(0.2*img3)+(0.2*img4)+(0.2*img5)
img_p = img_p.astype('uint8')

plt.imshow(cv2.cvtColor(img_p, cv2.COLOR_BGR2RGB))
```

```python
def gaussian_noise(img):
    mean = 0
    std_dev = 50
    noise = np.random.normal(mean, std_dev, img_p.shape)
    img_noise = img + noise
    img_noise = np.clip(img_noise, 0, 255).astype('uint8')

    _, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(25,10))
    ax1.set_title('Normal image')
    ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

    ax2.set_title('Noisy Image')
    ax2.imshow(cv2.cvtColor(img_noise, cv2.COLOR_BGR2RGB))

    ax3.set_title('Histogram')
    ax3.hist(noise.ravel(), bins=100, density=True, color='r')
    ax3.set_title('Histogram')
    ax3.set_xlabel('Intensity')
    ax3.set_ylabel('Frequency')

gaussian_noise(img_p)
```
