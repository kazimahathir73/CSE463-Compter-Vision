```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

```python
img1 = cv2.imread("./pizza.jpg")
gray1 = cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
gray1 = np.float32(gray1)

points1 = cv2.cornerHarris(gray1, 4, 7, 0.02)
points1 = cv2.dilate(points1, None)

pointer1 = 0.01 * points1.max()
img1[points1 > pointer1] = [0, 255, 0]

plt.imshow(img1, cmap='gray')
```

1. It is not detecting the pepperoni corectly. It is showing some points but not all the edges
2. One side if crust is correctly detected but other 2 side is not detected.
3. It can't detect chesse texture. Cause the area is has smooth intensity whoch leads to low gradient.

The Harris Corner Detection algorithm struggles with circular shapes and irregular toppings because it is designed to detect sharp, orthogonal intensity changes. It doesn't have smooth curves or irregular edges. Circular shapes have smooth edges with no sharp changes, and irregular toppings don't have clear corners, so the algorithm may miss them or detect incorrectly.

```python
img2 = cv2.imread("./cake.jpg")
gray2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
gray2 = np.float32(gray2)

points2 = cv2.cornerHarris(gray2, 4, 7, 0.02)
points2 = cv2.dilate(points2, None)

pointer2 = 0.01 * points2.max()
img2[points2 > pointer2] = [0, 255, 0]


plt.imshow(img2, cmap='gray')
```

The cake makes harris more happy. It give more accurate detect prediction then pizza slice.
